```python
import numpy as np
import matplotlib.pyplot as plt
from System_obc import *
```

```python
# Initialize ladder with s=1, L=8, electric potential p=0.5
s = 1 # only s=1 has full functionality
L = 8 
T = 10 # maximal time
steps = 10*T # time steps
system1 = System(s, L, potential=0.5)
# Initialize zero-mode scar in the +1 charge conjugation sector
scar = Scar(s, L, 3, 'even')
```

```python
fidelity = system1.time_evolve_vector_fidelity(T, steps, scar=scar)
time = np.linspace(0, T, steps+1)
plt.plot(time, fidelity)
plt.xlabel(r"$t$")
plt.ylabel(r"$F(t)$")
plt.title("Fidelity for electric potential with strength 0.5")
plt.show()
```
